# Medical Student Mental Health

## Problem Statement

### Goal

Exploring the correlation between gender, job satisfaction and empathy in medical students Investigating how language spoken by medical students relates to their mental health and burnout levels Predicting mental health and burnout levels in medical students based on different demographic variables such as age, sex, year of study and health status

### Dataset Description

This dataset contains information on the mental health, empathy and burnout of medical students in Switzerland. It includes data about demographic factors such as age, sex and language spoken as well as internal measures such as job satisfaction, psychological distress, education grades, self-reported health status and empathy rating scales. The objective of this dataset is to gain insight into the wellbeing of medical students in order to promote better policies by understanding relationships between these variables. Thus, this dataset can be used to identify how different variables may be related to the mental health of medical students.

Dataset: [Click Here](https://www.kaggle.com/datasets/thedevastator/medical-student-mental-health)

```python
#
# Plot Correlation Heatmap
#
plt.figure(figsize=(15,10))
cor = df_data.corr()
sns.heatmap(cor,annot=True,linewidth=.5,cmap='magma')
plt.title('Correlation Heatmap', fontsize = 30)
plt.show()
```

### Understanding Data Distribution

```python
#
# Identifying Outliers using Boxplot 
#
data_cols = ['age', 'psyt', 'jspe', 'qcae_cog', 'qcae_aff', 'amsp', 'erec_mean', 'cesd', 'stai_t', 'mbi_ex', 'mbi_cy', 'mbi_ea']
fig, axs = plt.subplots(3, 4, figsize=(16, 10))
for i, column in enumerate(data_cols):
    row = i // 4
    col = i % 4
    sns.boxplot(df_data[column], ax=axs[row, col])
    axs[row, col].set_xlabel(column)
plt.show()
```

```python
#
# Understanding data distribution using Histogram for - year
#
sns.histplot(df_data['year'])
```

```python
#
# Understanding data distribution using Histogram for - health
#

sns.histplot(df_data['health'])
```

```python
#
# Understanding data distribution using Histogram
#
data_cols = ['age', 'psyt', 'jspe', 'qcae_cog', 'qcae_aff', 'amsp', 'erec_mean', 'cesd', 'stai_t', 'mbi_ex', 'mbi_cy', 'mbi_ea']
fig, axs = plt.subplots(3, 4, figsize=(16, 10))
for i, column in enumerate(data_cols):
    row = i // 4
    col = i % 4
    sns.histplot(data = df_data, x=column, bins=len(df_data[column].unique()), ax=axs[row, col], kde=True, hue='health')
    axs[row, col].set_xlabel(column)
plt.show()
```

```python
#
# Understanding data distribution using Histogram
#
data_cols = ['age', 'psyt', 'jspe', 'qcae_cog', 'qcae_aff', 'amsp', 'erec_mean', 'cesd', 'stai_t', 'mbi_ex', 'mbi_cy', 'mbi_ea']
fig, axs = plt.subplots(3, 4, figsize=(16, 10))
for i, column in enumerate(data_cols):
    row = i // 4
    col = i % 4
    sns.boxplot(data = df_data, x=column, ax=axs[row, col], y='health')
    axs[row, col].set_xlabel(column)
plt.show()
```
